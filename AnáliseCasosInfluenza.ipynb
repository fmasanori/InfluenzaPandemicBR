```python
import pandas as pd
from pandas import *
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/Influenza_UF_FaixaEtaria.csv', sep = ";", encoding='latin-1')
data.head()
```

```python
uf = data.groupby('UF de notificação').sum()['Total'].sort_values(ascending=False)
uf
```

```python
plt.style.use("ggplot")
uf.loc['41 Paraná':'16 Amapá'].plot(kind='bar', subplots=True, label="Total de casos de Influenza por estado de 2009 e 2010" ,figsize=(10, 5), color='#ff0099')
plt.xticks(rotation=90)
```

```python
plt.style.use("ggplot")
uf.loc['41 Paraná':'43 Rio Grande do Sul'].plot(kind='bar', subplots=True, label="Os 5 estados com mais casos de Influenza de 2009 e 2010" ,figsize=(9, 5), color='#ff0099')
plt.xticks(rotation=360)
```

```python
labels = '41 Paraná','35 São Paulo','33 Rio de Janeiro','31 Minas Gerais','43 Rio Grande do Sul'
sizes = uf.loc['41 Paraná':'43 Rio Grande do Sul']
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels,autopct='%1.1f%%',shadow=False,startangle=90)
ax1.legend(title="Estados",
          loc="center left",
          bbox_to_anchor=(1.1, 0, 0.5, 1))

ax1.set_title("Os 5 estados com mais casos de Influenza de 2009 e 2010 em porcentagem")
ax1.axis('equal')
plt.show()
```

```python
obitos = data.groupby('UF de notificação').sum()['Obitos'].sort_values(ascending=False)
obitos
```

```python
plt.style.use("ggplot")
obitos.loc['35 São Paulo':'27 Alagoas'].plot(kind='bar', subplots=True, label="Total Óbitos por Influenza (2009 a 2010)" ,figsize=(15, 7), color='#6e0dd0')
plt.xticks(rotation=90)
```
